```python
import nltk
from nltk import word_tokenize, pos_tag, ne_chunk
from nltk.tree import Tree
from nltk.probability import FreqDist
from nltk.util import bigrams, trigrams

# Download required datasets
nltk.download('punkt')
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')

# =========================
# INPUT
# =========================
text = input("Enter News Article:\n")

# =========================
# TOKENIZATION + POS TAGGING
# =========================
tokens = word_tokenize(text)
tagged = pos_tag(tokens)

print("\n--- POS TAGS ---")
print(tagged)

# =========================
# SIMPLE PARSING
# =========================
subject = ""
verb = ""
obj = ""

for word, tag in tagged:
    if tag.startswith("NN") and subject == "":
        subject = word
    elif tag.startswith("VB") and verb == "":
        verb = word
    elif tag.startswith("NN") and subject != "" and verb != "":
        obj = word

print("\n--- PARSING ---")
print("Subject:", subject)
print("Verb:", verb)
print("Object:", obj)

# =========================
# NAMED ENTITY RECOGNITION
# =========================
print("\n--- NAMED ENTITIES ---")

tree = ne_chunk(tagged)
print(tree)

# =========================
# WORD FREQUENCY
# =========================
words = [w.lower() for w in tokens if w.isalpha()]
fdist = FreqDist(words)

print("\n--- WORD FREQUENCY ---")
for w, c in fdist.most_common(10):
    print(w, ":", c)

# =========================
# PROBABILITY
# =========================
total_words = len(words)

print("\n--- PROBABILITY ---")
for w, c in fdist.most_common(5):
    print(f"P({w}) = {c}/{total_words} = {round(c/total_words, 2)}")

# =========================
# BIGRAMS
# =========================
print("\n--- BIGRAMS ---")
print(list(bigrams(words)))

# =========================
# TRIGRAMS
# =========================
print("\n--- TRIGRAMS ---")
print(list(trigrams(words)))
```

```python
import nltk
from nltk.classify import NaiveBayesClassifier
from nltk.classify.util import accuracy
from nltk.corpus import wordnet as wn

nltk.download('wordnet')

# =========================
# INPUT
# =========================
text = input("Enter News Article:\n")

tokens = nltk.word_tokenize(text)

# =========================
# CLASSIFICATION MODEL
# =========================
train_data = [
    ({"excellent": True, "friendly": True}, "Positive"),
    ({"outstanding": True, "good": True}, "Positive"),
    ({"poor": True, "bad": True}, "Negative"),
    ({"terrible": True, "rude": True}, "Negative")
]

classifier = NaiveBayesClassifier.train(train_data)

features = {
    "excellent": "excellent" in text.lower(),
    "friendly": "friendly" in text.lower(),
    "outstanding": "outstanding" in text.lower()
}

category = classifier.classify(features)

print("\n--- CLASSIFICATION ---")
print("Category:", category)

# =========================
# ACCURACY
# =========================
test_data = [
    ({"excellent": True, "friendly": True}, "Positive"),
    ({"poor": True, "bad": True}, "Negative")
]

print("\n--- ACCURACY ---")
print("Accuracy:", accuracy(classifier, test_data))

# =========================
# SEMANTIC ANALYSIS
# =========================
print("\n--- SEMANTIC MEANING (WORDNET) ---")

for syn in wn.synsets("technology")[:3]:
    print("-", syn.definition())
```
