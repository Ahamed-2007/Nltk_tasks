```python
import nltk

# ================= DOWNLOAD REQUIRED NLTK DATA =================
nltk.download('punkt_tab')
nltk.download('stopwords')
nltk.download('wordnet')
nltk.download('omw-1.4')
nltk.download('averaged_perceptron_tagger_eng')

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.probability import FreqDist

# ================= SAMPLE TEXT =================
text = """
John is a Python Developer with experience in Machine Learning,
Data Analysis, SQL, and Web Development.
"""

# ================= TOKENIZATION =================
tokens = word_tokenize(text)

# ================= STOPWORD REMOVAL =================
stop_words = set(stopwords.words('english'))

filtered_words = [
    word.lower()
    for word in tokens
    if word.isalpha() and word.lower() not in stop_words
]

# ================= STEMMING =================
ps = PorterStemmer()
stemmed_words = [ps.stem(word) for word in filtered_words]

# ================= LEMMATIZATION =================
lemmatizer = WordNetLemmatizer()
lemmatized_words = [
    lemmatizer.lemmatize(word)
    for word in filtered_words
]

# ================= POS TAGGING =================
pos_tags = nltk.pos_tag(lemmatized_words)

# ================= FREQUENCY DISTRIBUTION =================
fdist = FreqDist(lemmatized_words)

# ================= OUTPUT =================
print("\n================ TOKENIZATION ================\n")
print(tokens)

print("\n================ STOPWORD REMOVAL ================\n")
print(filtered_words)

print("\n================ STEMMING ================\n")
print(stemmed_words)

print("\n================ LEMMATIZATION ================\n")
print(lemmatized_words)

print("\n================ POS TAGGING ================\n")
print(pos_tags)

print("\n================ FREQUENCY DISTRIBUTION ================\n")
print(fdist.most_common())
```
